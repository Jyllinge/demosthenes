```python
import pickle
import pandas as pd
import numpy as np
import torch
import transformers
import bitsandbytes as bnb
import networkx as nx

random_state = 1
```

```python
def euclid(list1, list2):
    return sum((p-q)**2 for p, q in zip(list1, list2)) ** .5
```

```python
with open('df_annotations_embeddings.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
df_train = df[df['Split'] != 2]
df_train = df_train.dropna(subset = ['Scheme'])
df_train
```

```python
distances = []
for j in range(len(df_train)):
    newlist = []
    for i in range(len(df_train)):
        newlist.append(euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']))
    distances.append(newlist)
    print(j)
```

(779*50)**.5 = 197.358

```python
closenodes = []
for el in distances:
    count = 0
    for newel in el:
        if newel <= 29.02:      #set manually, with 46.14 on average the extracted subgraph has (N*50)**0.5 nodes, with 30.86 the nodes are connected to 1% of other nodes
                                # with 41.23, half nodes                                                                 with 29.02   .5% other nodes
                                # with 54.04, double nodes                                                               with 33.06   2% other nodes
            count += 1
    closenodes.append(count)

print(sum(closenodes)/len(closenodes))  #average size of extracted subgraph
```

```python
Training_graph = nx.Graph() #Resoluti

for i in range(len(df_train)):
    print('adding node',i)
    Training_graph.add_node(i)
    for j in range(i):
        if euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']) <= 30.86:      #così in media ogni nodo è collegato all'1%
            Training_graph.add_edge(i,j)
```

```python
pickle.dump(Training_graph, open('task3_all_.01.pickle', 'wb'))
```

```python
nx.draw_networkx(Training_graph, with_labels = False)
```
