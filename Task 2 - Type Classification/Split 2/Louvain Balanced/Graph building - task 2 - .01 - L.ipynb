```python
import pickle
import pandas as pd
import numpy as np
import torch
import transformers
import bitsandbytes as bnb
import networkx as nx

random_state = 1
```

```python
def euclid(list1, list2):
    return sum((p-q)**2 for p, q in zip(list1, list2)) ** .5
```

```python
with open('df_annotations_embeddings.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
def contains_str(value, string):
    return string in value
```

```python
df_train = df[df['Split'] != 2]
df_train = df_train.dropna(subset = ['Type'])
df_train = df_train[df_train['Type'].apply(lambda x: contains_str(x, 'L'))]
df_train
#df_train[df_train['Type'] == ['F', 'L']]
```

```python
distances = []
for j in range(len(df_train)):
    newlist = []
    for i in range(len(df_train)):
        newlist.append(euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']))
    distances.append(newlist)
    print(j)
```

(773*25)**.5 = 139.014

```python
closenodes = []
for el in distances:
    count = 0
    for newel in el:
        if newel <= 33.01:      #set manually, with 43.52 on average the extracted subgraph has (N*25)**0.5 nodes, with 30.83 the nodes are connected to 1% of other nodes
                                # with 39.34, half nodes                                                                 with 28.97   .5% other nodes
                                # with 49.58, double nodes                                                               with 33.02   2% other nodes
            count += 1
    closenodes.append(count)

print(sum(closenodes)/len(closenodes))  #average size of extracted subgraph
print(min(closenodes))                  #min degree
print(max(closenodes))                  #max degree
```

```python
Training_graph = nx.Graph() #Resoluti

for i in range(len(df_train)):
    print('adding node',i)
    Training_graph.add_node(i)
    for j in range(i):
        if euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']) <= 30.83:      #così in media ogni nodo è collegato all'1%
            Training_graph.add_edge(i,j)
```

```python
pickle.dump(Training_graph, open('task2_.01_L.pickle', 'wb'))
```

```python
nx.draw_networkx(Training_graph, with_labels = False)
```
