```python
import pickle
import pandas as pd
import numpy as np
import torch
import transformers
import bitsandbytes as bnb
import networkx as nx

random_state = 1
```

```python
def euclid(list1, list2):
    return sum((p-q)**2 for p, q in zip(list1, list2)) ** .5
```

```python
with open('df_annotations_embeddings.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
df_train = df[df['Split'] != 2]
df_train = df_train[df_train['Name'] == 'prem']
df_train
```

```python
df_test = df[df['Split'] == 2]
df_test
```

```python
distances = []
for j in range(len(df_train)):
    newlist = []
    for i in range(len(df_train)):
        newlist.append(euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']))
    distances.append(newlist)
    print(j)
```

sqrt(2029*25) = 225.222

```python
closenodes = []
for el in distances:
    count = 0
    for newel in el:
        if newel <= 35.2:      #set manually, with 41.93 on average the extracted subgraph has (N*25)**0.5 nodes, with 33.21 the nodes are connected to 1% of other nodes, on average
                                #46.26 double nodes                                                                 31.46, .5% of other nodes                                                      
                                #38.79 half nodes                                                                   35.20, 2% of other nodes
            count += 1
    closenodes.append(count)

print(sum(closenodes)/len(closenodes))  #average size of extracted subgraph
print(min(closenodes))                  #min degree
print(max(closenodes))                  #max degree
```

```python
Training_graph = nx.Graph()

for i in range(len(df_train)):
    print('adding node',i)
    Training_graph.add_node(i)
    for j in range(i):
        if euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']) <= 33.21:      #così in media ogni nodo è collegato all'1% degli altri nodi, in media
            Training_graph.add_edge(i,j)
```

```python
pickle.dump(Training_graph, open('premgraph_task1_.01.pickle', 'wb'))
```

```python
nx.draw_networkx(Training_graph, with_labels = False)
```
