```python
import pickle
import pandas as pd
import numpy as np
import torch
import transformers
import bitsandbytes as bnb
import networkx as nx

random_state = 1
```

```python
def euclid(list1, list2):
    return sum((p-q)**2 for p, q in zip(list1, list2)) ** .5
```

```python
with open('df_annotations_embeddings.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
df_train = df[df['Split'] != 2]
df_train
```

```python
df_test = df[df['Split'] == 2]
df_test
```

```python
distances = []
for j in range(len(df_train)):
    newlist = []
    for i in range(len(df_train)):
        newlist.append(euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']))
    distances.append(newlist)
    print(j)
```

(2167*50)**.5 = 329.166

```python
closenodes = []
for el in distances:
    count = 0
    for newel in el:
        if newel <= 35.48:      #set manually, with 44.18 on average the extracted subgraph has (N*50)**0.5 nodes, with 33.43 the nodes are connected to 1% of other nodes
                                # with 40.49, half nodes                                                                 with 31.61   .5% other nodes
                                # with 49.55, double nodes                                                               with 35.48   2% other nodes
            count += 1
    closenodes.append(count)

print(sum(closenodes)/len(closenodes))  #average size of extracted subgraph
print(min(closenodes))                  #min degree
print(max(closenodes))                  #max degree
```

```python
Training_graph = nx.Graph()

for i in range(len(df_train)):
    print('adding node',i)
    Training_graph.add_node(i)
    for j in range(i):
        if euclid(df_train.iloc[j]['embeddings'], df_train.iloc[i]['embeddings']) <= 33.43:      #così in media ogni nodo è collegato all'1%
            Training_graph.add_edge(i,j)
```

```python
pickle.dump(Training_graph, open('task1_.01.pickle', 'wb'))
```

```python
nx.draw_networkx(Training_graph, with_labels = False)
```
